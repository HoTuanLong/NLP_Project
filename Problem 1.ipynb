```python
import math
```

```python
# convert data from txt file to dict
d = {}
i = 0
with open("W2V_150.txt") as f:
    for line in f:
        i += 1
        if i >= 3:
            a = line.split()
            key = a[0]
            a.pop(0)
            val = a
            d[str(key)] = val
            for i in range(0, len(val)):
                val[i] = float(val[i])
```

```python
def cosine_similarity(v1,v2):
    sumxx, sumxy, sumyy = 0, 0, 0
    for i in range(len(v1)):
        x = v1[i]; y = v2[i]
        sumxx += x*x
        sumyy += y*y
        sumxy += x*y
    return sumxy/math.sqrt(sumxx*sumyy)
```

```python
i = 0
with open("ViSim-400.txt") as f:
    for line in f:
        i += 1
        if i > 1:
            a = line.split()
            if a[0] in d.keys() and a[1] in d.keys():
                v1 = d.get(a[0])        
                v2 = d.get(a[1])
                print(cosine_similarity(v1, v2))
```
